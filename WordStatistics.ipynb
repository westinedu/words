```python
from docx import Document
import re

def count_words_in_docx(docx_path):
    # Open the document
    doc = Document(docx_path)
    
    # Initialize word count
    word_count = 0
    
    # Iterate over each paragraph in the document
    for paragraph in doc.paragraphs:
        # Split the paragraph text into words and count them
        words = paragraph.text.split()
      
        word_count += len(words)
    
    return word_count

def get_word_list_from_docx(docx_path):
    # Open the document
    doc = Document(docx_path)
    
    # Initialize a list to hold all words
    word_list = []
    
    # Iterate over each paragraph in the document
    for paragraph in doc.paragraphs:
        # Split the paragraph text into words and extend the word list
        words = paragraph.text.split()
        word_list.extend(words)
    
    return word_list

def count_unique_words(text):  
    # 1. 分词  
    words = re.findall(r'\b\w+\b', text.lower())  # 使用正则表达式匹配单词，并转换为小写  
  
    # 2. 去重  
    unique_words = set(words)  
  
    # 3. 计数  
    return len(unique_words) 

def get_unique_words(docx_path):
    # Open the document
    doc = Document(docx_path)
    
    # Initialize a set to hold unique words
    unique_words = set()
    
    # Regular expression to match words
    word_regex = re.compile(r'\b\w+\b')
    
    # Iterate over each paragraph in the document
    for paragraph in doc.paragraphs:
        # Find all words in the paragraph text
        words = word_regex.findall(paragraph.text.lower())
        # Add words to the set
        unique_words.update(words)
    
    # Return the count of unique words
    return sorted(unique_words)

def write_words_to_docx(words, output_path):
    # Create a new Document
    doc = Document()
    
    # Add each word to the document
    for word in words:
        doc.add_paragraph(word)
    
    # Save the document
    doc.save(output_path)

def write_words_to_txt(words, output_path):
    # Open the text file for writing
    # with open(output_path, 'w') as file:
    # Open the text file for writing with UTF-8 encoding
    with open(output_path, 'w', encoding='utf-8') as file:
        # Write each word on a new line
        for word in words:
            file.write(word + '\n')


# Example usage
docx_path = 'J:\雅思\雅思作文\IELTS 小作文.docx'
output_docx_path = 'J:\雅思\雅思作文\小作文 words.docx'
output_txt_path = 'J:\雅思\雅思作文\小作文 words.txt'



total_words = count_words_in_docx(docx_path)
word_list = get_word_list_from_docx(docx_path)
print(f'Total word : {word_list}')
unique_words = get_unique_words(docx_path)
print(f'Total word count: {total_words}')
print(f'unique words list: { unique_words}')
print(f'Total unique word count: {len(unique_words)}')

# Write the unique words to the output document
write_words_to_docx(unique_words, output_docx_path)
write_words_to_txt(unique_words, output_txt_path)


```
